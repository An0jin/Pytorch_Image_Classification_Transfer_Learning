```python
import pandas as pd
df=pd.read_csv("result.csv",index_col="epoch")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(2, 5, figsize=(25, 10))
# 2. 차트에 표시할 칼럼 리스트 정의
cols = [
    'train_accuracy_score', 'train_loss', 'train_precision_score', 'train_recall_score', 'train_f1_score',
    'valid_accuracy_score', 'valid_loss', 'valid_precision_score', 'valid_recall_score', 'valid_f1_score'
]

# 3. axes를 1차원으로 평탄화하여 반복문 적용
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.lineplot(data=df, x='epoch', y=col, ax=axes[i])
    axes[i].set_title(col)
    axes[i].grid(True)

# 4. 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('results.png', dpi=300, bbox_inches='tight')
plt.show()
```
